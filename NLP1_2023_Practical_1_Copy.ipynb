Lexica such as this can be used to solve
the classification task without using Machine Learning. For example, one might look up every word $w_1 ... w_n$ in a document, and compute a **binary score**
$S_{binary}$ by counting how many words have a positive or a
negative label in the sentiment lexicon $SLex$.

$$S_{binary}(w_1 w_2 ... w_n) = \sum_{i = 1}^{n}\text{sign}(SLex\big[w_i\big])$$

where $\text{sign}(SLex\big[w_i\big])$ refers to the polarity of $w_i$.

**Threshold.** On average, there are more positive than negative words per review (~7.13 more positive than negative per review) to take this bias into account you should use a threshold of **8** (roughly the bias itself) to make it harder to classify as positive.

$$
\text{classify}(S_{binary}(w_1 w_2 ... w_n)) = \bigg\{\begin{array}{ll}
        \text{positive} & \text{if } S_{binary}(w_1w_2...w_n) > threshold\\
        \text{negative} & \text{otherwise}
        \end{array}
$$


#### (Q1.1) Implement this approach and report its classification accuracy. (1 pt)

```python
# YOUR CODE HERE
lexicon_file = 'sent_lexicon.txt'
reviews_file = 'reviews.json'

lexicon = {}
with open(lexicon_file, mode="r", encoding="utf-8") as lx:
    for line in lx:
        elements = line.strip().split()
        word = elements[2].split('=')[-1].lower()
        polarity = elements[-1].split('=')[-1]
        lexicon[word] = polarity
# print(lexicon)


def s_binary(tokens):
    sb = 0
    count = 0 
    for i in tokens:
        word_lower = i.lower()
        if word_lower in lexicon:
            count+=1
            if lexicon[word_lower] == 'positive':
                sb += 1
            if lexicon[word_lower] == 'negative':
                sb -= 1

    # print(sb)
    return sb


with open(reviews_file, mode="r", encoding="utf-8") as f:
    reviews = json.load(f)
    
token_result = []
threshold = 8 

for i, r in enumerate(reviews):
    content = r["content"]
    words = []
    for sentence in content:
        for token, pos_tag in sentence:
            words.append(token)
    
    # print(words)
    # break
    binary_score = s_binary(words)

    # Classify the review based on the threshold
    if binary_score > threshold:
        classification = 'POS'
    else:
        classification = 'NEG'

    if r["sentiment"] == classification:
        token_result.append(1)
    else:
        token_result.append(0)
```

As the sentiment lexicon also has information about the **magnitude** of
sentiment (e.g., *“excellent"* has the same sentiment _polarity_ as *“good"* but it has a higher magnitude), we can take a more fine-grained approach by adding up all
sentiment scores, and deciding the polarity of the movie review using
the sign of the weighted score $S_{weighted}$.

$$S_{weighted}(w_1w_2...w_n) = \sum_{i = 1}^{n}SLex\big[w_i\big]$$


Make sure you define an appropriate threshold for this approach.

#### (Q1.2) Now incorporate magnitude information and report the classification accuracy. Don't forget to use the threshold. (1pt)

```python
# YOUR CODE HERE
lex = 'sent_lexicon.txt'
reviews_file = 'reviews.json'

lexicon = {}
with open(lex, mode="r", encoding="utf-8") as lx:
    for line in lx:
        elements = line.strip().split()
        word = elements[2].split('=')[-1].lower()
        polarity = elements[-1].split('=')[-1]
        magnitude = elements[0].split('=')[-1]
        lexicon[word] = {'polarity': polarity, 'magnitude': magnitude}

def s_weighted(tokens):
    sw = 0
    for i in tokens:
        word_lower = i.lower()
        if word_lower in lexicon:
            polarity = lexicon[word_lower]['polarity']
            magnitude = lexicon[word_lower]['magnitude'] 
            
            if magnitude == 'strongsubj':
                weight = 2.0  
            elif magnitude == 'weaksubj':
                weight = 1.0  
            
            if polarity == 'positive':
                sw += weight
            elif polarity == 'negative':
                sw -= weight
                
    return sw
    
token_results = []
threshold = 9  

#hihg wights, high threshold

for i, r in enumerate(reviews):
    content = r["content"]
    words = []
    for sentence in content:
        for token, pos_tag in sentence:
            words.append(token)
  
    weighted_score = s_weighted(words)
    
    # for debugging
    # print(f"Review {i}: Weighted Score: {weighted_score}, Threshold: {threshold}")

    # Decide the polarity of the review based on the sign of the weighted score
    if weighted_score > threshold:
        sentiment = 'POS'
    else:
        sentiment = 'NEG'

    if r["sentiment"] == sentiment:
        token_results.append(1)
    else:
        token_results.append(0)
```

## Writing your own classifier (4pts)

Write your own code to implement the Naive Bayes (NB) classifier. As
a reminder, the Naive Bayes classifier works according to the following
equation:
$$\hat{c} = \operatorname*{arg\,max}_{c \in C} P(c|\bar{f}) = \operatorname*{arg\,max}_{c \in C} P(c)\prod^n_{i=1} P(f_i|c)$$
where $C = \{ \text{POS}, \text{NEG} \}$ is the set of possible classes,
$\hat{c} \in C$ is the most probable class, and $\bar{f}$ is the feature
vector. Remember that we use the log of these probabilities when making
a prediction:
$$\hat{c} = \operatorname*{arg\,max}_{c \in C} \Big\{\log P(c) + \sum^n_{i=1} \log P(f_i|c)\Big\}$$

#### (Q2.2) Train your classifier on (positive and negative) reviews with cv-value 000-899, and test it on the remaining (positive and negative) reviews cv900–cv999.  Report results using classification accuracy as your evaluation metric. Your  features are the word vocabulary. The value of a feature is the count of that feature (word) in the document. (2pts)


```python
with open('reviews.json', 'r') as file:
    reviews = json.load(file)


def tokenize(text):
    """
    Gets a text string/ sentence as a parameter and returns a list with 
    lower cased tokens from the sentences.
    :param text: string
    :returns: list with lowercased tokens
    """
    words = [token.lower() for sentence in text for token, _ in sentence]
    return words


def train_naive_bayes(train_data, kappa=None):
    """
    recieves a training data and optional smoothing parameter which is used to 
    train the naive bayes classifier algorithm. It initliazes and populates the word frequencies
    and vocabularies with sentiments and their respective quantity per review. It calculates the prior
    prbability per sentiment {POS or NEG}. Lastly, conditional probabilities are calculated with(out) the
    smoothing paramter and computed using the word freqency values and sentiment values. 
    :param train_data: list containing reviews
    :param kappa: smoothing factor
    :returns: vocabulary with tokenized words, prior probabilities,conditional probabilities and word frequency
    dictionary per sentiment.
    """
    vocab = set()
    class_count = {'POS': 0, 'NEG': 0}
    word_freq = {'POS': {}, 'NEG': {}}

    # Initializing frequency dictionary and vocabulary
    for sentiment in ['POS', 'NEG']:
        for review in train_data:
            if review['sentiment'] == sentiment:
                class_count[sentiment] += 1
                words = tokenize(review['content'])

                for word in words:
                    if word not in word_freq[sentiment]:
                        word_freq[sentiment][word] = 0
                    
                    word_freq[sentiment][word] += 1
                    vocab.add(word)

    # Calculating prior probabilities
    prior = {}
    for sentiment in class_count:
        prior[sentiment] = math.log(class_count[sentiment] / len(train_data))
    
    # Calculating conditional probabilities 
    cond_prob = {}
    for word in vocab:
        if word_freq['POS'].get(word, 0) > 0 and word_freq['NEG'].get(word, 0) > 0:
            cond_prob[word] = {}
            for sentiment in ['POS', 'NEG']:

                if kappa is not None: # Applying Laplace smoothing
                    freq = word_freq[sentiment].get(word, 0) + kappa 
                    total_freq = sum(word_freq[sentiment].values()) + kappa * len(vocab)

                else: # Case when No smoothing is applied
                    freq = word_freq[sentiment].get(word, 0)
                    total_freq = sum(word_freq[sentiment].values())
                    
                cond_prob[word][sentiment] = math.log(freq / total_freq) #storing conditional probability per word/sentiment

    print(cond_prob)
    return vocab, prior, cond_prob, word_freq


def apply_naive_bayes(vocab, prior, condprob, word_freq, review,kappa=None):
    """
    Uses the return paramters calculated during the training of the NB classifer to 
    apply it to an extracted review from the test data set and ultimately return a predicted sentiment.
    :param vocab: set with unique words
    :param prior: dictionary storing prior probabilty values per sentiment 
    :param condprob: dictionary storing condtional probabilty values per word per sentiment 
    :param word_freq: dictionary storing the frequencies of the sentiment values from reviews/extracted words
    :param review: a review from the reviews dataset extracted from test dataset
    :param kappa: smoothing factor
    :returns: prediction of the naive bayes classifier
    """
    tokens = tokenize(review['content']) #lowercasing text 
    score = {'POS': prior['POS'], 'NEG': prior['NEG']} #filling it initillay with the prior probabilites extracted during training

    for token in tokens:
        if token in vocab:
            for sentiment in ['POS', 'NEG']:
                if kappa is not None:
                # Checking if token is in conditional probability and then adding score with Laplace smoothing factor
                    if token in condprob and sentiment in condprob[token]:
                        score[sentiment] += condprob[token][sentiment]
                    else:
                    # Applying Laplace smoothing if the token was not seen in the training data for this class
                        score[sentiment] += math.log(kappa / (sum(word_freq[sentiment].values()) + kappa * len(vocab)))
                else:
                # If no smoothing is applied and the token is in the conditional probabilities
                    if token in condprob:
                        score[sentiment] += condprob[token].get(sentiment, 0)
                

    if score['POS'] > score['NEG']:
        prediction = 'POS'
    else:
        prediction = 'NEG'
    
    return prediction




def calculate_accuracy(vocab, prior, condprob, word_freq, test_data, kappa=None):
    """
    Calculates the accuracy of the Naive Bayes classifier by using the return paramters during
    the training phase of NB classifer and comparing the prediction of NB classifer with the predictions
    of held-out set.   
    :param vocab: set with unique words
    :param prior: dictionary storing prior probabilty values per sentiment 
    :param condprob: dictionary storing condtional probabilty values per word per sentiment 
    :param word_freq: dictionary storing the frequencies of the sentiment values from reviews/extracted words
    :param test_data: list storing the testing dataset retirved from reviews file
    :param kappa: smoothing factor
    :returns: accuracy score of the NB classifer
    """
    correctly_classified = 0
    for review in test_data:
        prediction = apply_naive_bayes(vocab, prior, condprob,word_freq, review,kappa)
        if prediction == review['sentiment']:
            correctly_classified += 1
    return correctly_classified / len(test_data)
```

```python
# Separating the reviews into training and testing sets
train_data = [review for review in reviews if int(review['cv']) < 900]
test_data = [review for review in reviews if int(review['cv']) >= 900]

#Set kappa to None to not use smoothing or to 1 for Laplace smoothing (useful for Q2.4)
kappa_value = None  

#training phase 
vocab, prior, condprob, word_freq = train_naive_bayes(train_data, kappa=kappa_value)

#testing phase
accuracy = calculate_accuracy(vocab, prior, condprob, word_freq, test_data,kappa=kappa_value)

print(f"Classification Accuracy when (kappa={kappa_value}): {accuracy * 100:.2f}%")
```

## Smoothing (1pt)

As mentioned above, the presence of words in the test dataset that
have not been seen during training can cause probabilities in the Naive
Bayes classifier to be $0$, thus making that particular test instance
undecidable. The standard way to mitigate this effect (as well as to
give more clout to rare words) is to use smoothing, in which the
probability fraction
$\frac{\text{count}(w_i, c)}{\sum\limits_{w\in V} \text{count}(w, c)}$ for a word
$w_i$ becomes
$$\frac{\text{count}(w_i, c) + \text{smoothing}(w_i)}{\sum\limits_{w\in V} \text{count}(w, c) + \sum\limits_{w \in V} \text{smoothing}(w)}$$

## Cross-Validation (1.5pts)

A serious danger in using Machine Learning on small datasets, with many
iterations of slightly different versions of the algorithms, is ending up with Type III errors, also called the “testing hypotheses
suggested by the data” errors. This type of error occurs when we make
repeated improvements to our classifiers by playing with features and
their processing, but we don’t get a fresh, never-before seen test
dataset every time. Thus, we risk developing a classifier that gets better
and better on our data, but only gets worse at generalizing to new, unseen data. In other words, we risk developping a classifier that overfits.

A simple method to guard against Type III errors is to use
Cross-Validation. In **N-fold Cross-Validation**, we divide the data into N
distinct chunks, or folds. Then, we repeat the experiment N times: each
time holding out one of the folds for testing, training our classifier
on the remaining N - 1 data folds, and reporting performance on the
held-out fold. We can use different strategies for dividing the data:

-   Consecutive splitting:
  - cv000–cv099 = Split 1
  - cv100–cv199 = Split 2
  - etc.
  
-   Round-robin splitting (mod 10):
  - cv000, cv010, cv020, … = Split 1
  - cv001, cv011, cv021, … = Split 2
  - etc.

-   Random sampling/splitting
  - Not used here (but you may choose to split this way in a non-educational situation)
